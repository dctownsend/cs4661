# A. Read the dataset file “Credit.csv” (from github using the following command), and assign it to a Pandas DataFrame: df  = pd.read_csv("https://github.com/mpourhoma/CS4661/raw/master/Credit.csv")

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression

import numpy as np
import pandas as pd
```

```python
#reading csv file and storing it directly in a dataframe
df = pd.read_csv("https://github.com/mpourhoma/CS4661/raw/master/Credit.csv")
```

# B. Check out the dataset. The “Credit” dataset includes the “balance” column (average credit card debt for a number of individuals) as target, as well as several features: age, cards (number of credit cards), education (years of education), income (in thousands of dollars), limit (credit limit), marital status, and rating (credit rating).

```python
#check dataset every 15 lines
df[0::15]
```

# C. Generate the feature matrix and target vector (target is “balance” in this dataset). Then, normalize (scale) the features (note: don’t normalize the target vector!). To normalize the data, you can simply use scale(X)  from sklearn.

```python
# create list of feature names from dataset:

feature_cols = ['Age', 'Cards', 'Education', 'Income', 'Limit', 'Married', 'Rating']

# use list to select the features DataFrame

X = df[feature_cols]

# print the first 5 rows
X.head()
```

```python
## select the last column from DataFrame

y = df['Balance']

#print first 5 rows

y.head()
```

```python
#normalizing the features
X = preprocessing.scale(X)

#printing the dataset
X
```

# D. Split the dataset into testing and training sets with the following parameters: test_size=0.24, random_state=9.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.24, random_state=9)

print(X_test)
print('\n')
print(y_test)
```

# E. Use Linear Regression to train a linear model on the training set. Check the coefficients of the linear regression model. Which feature is the most important? Which feature is the least important?

```python
# "my_linreg" is instantiated as an "object" of LinearRegression "class" 

my_linreg = LinearRegression()

my_linreg.fit(X_train, y_train)
```

```python
# print theta0 - using attribute "intercept_":
print(my_linreg.intercept_)

# print theta1 theta2 theta3 - using attribute "coef_":
print(my_linreg.coef_)
```

```python
# most important is rating
# least important is income
```

# F. Predict “balance” for the users in testing set. Then, compare the predicted balance with the actual balance by calculating and reporting the RMSE (as we saw in lab tutorial 4).

```python
# make predictions on the testing set
y_prediction = my_linreg.predict(X_test)

print(y_prediction)
```

```python
from sklearn import metrics
import numpy as np

# calculate "Mean Square Error" (MSE):
mse = metrics.mean_squared_error(y_test, y_prediction)

# use numpy sqrt - take the square root & calculate "Root Mean Square Error" (RMSE)
rmse = np.sqrt(mse)

print(rmse)
```

```python
# RMSE is 143.24649739642265 - this rmse is pretty high
```

# G. Now, use 10-fold Cross-Validation to evaluate the performance of a linear regression in predicting the balance. Thus, rather than splitting the dataset into testing and training, use Cross-Validation to evaluate the regression performance. What is the RMSE when you use cross validation?

```python
# importing the method:
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.model_selection import cross_val_score

# apply 10-fold cross validation with "linear regression":

# "my_linreg" is instantiated as an "object" of LinearRegression "class". 
my_linreg = LinearRegression()

mse_list = cross_val_score(my_linreg, X, y, cv=10, scoring='neg_mean_squared_error')

print(mse_list)
```

```python
# to calculate rmse, we have to make them positive
mse_list_positive = -mse_list

# using sqrt function to calculate rmse:
rmse_list = np.sqrt(mse_list_positive)
print(rmse_list)
```

```python
# calculate avg RMSE as final result of cross validation:
print(rmse_list.mean())
```

```python
# RMSE is 160.33198910744142 when we use cross validation.
```
